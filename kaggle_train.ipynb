## Step 2 — Import Datasets from Kaggle

### How to add datasets:
1. Click **"Add Data"** (right sidebar) → search for the dataset

### Datasets to add:

|---------|------------|-------------------|

| **Celeb-DF v2** | `Celeb-DF` | `Celeb-real/` + `Celeb-synthesis/` + `YouTube-real/` |

Set the paths below to match your mounted datasets.

````python
# ╔══════════════════════════════════════════════════════════════════╗
# ║             SET YOUR KAGGLE DATASET PATHS                       ║
# ╚══════════════════════════════════════════════════════════════════╝

FFPP_ROOT    = "/kaggle/input/faceforensics"   # ← CHANGE THIS
CELEBDF_ROOT = "/kaggle/input/celeb-df-v2"     # ← CHANGE THIS

# Set to False to skip a dataset you don't have
USE_FFPP     = True
USE_CELEBDF  = True

# ═══════════════════════════════════════════════════════════════════

# Auto-detect dataset structures
FFPP_IS_PRE_EXTRACTED = False
CELEBDF_IS_PRE_EXTRACTED = False

def inspect_dataset(name, root, raw_markers, split_marker="train"):
    """Check if a dataset exists and detect its structure."""
    if not os.path.isdir(root):
        print(f"❌ {name} NOT found at: {root}")
        return None  # not found
    
    print(f"✅ {name} found at: {root}")
    for item in sorted(os.listdir(root)):
        full = os.path.join(root, item)
        marker = "📁" if os.path.isdir(full) else "📄"
        print(f"   {marker} {item}")
    
    has_raw = all(os.path.isdir(os.path.join(root, m)) for m in raw_markers)
    has_splits = os.path.isdir(os.path.join(root, split_marker))
    
    if has_raw:
        print(f"   🎬 Structure: Raw videos")
        return False  # not pre-extracted
    elif has_splits:
        print(f"   🖼️ Structure: Pre-extracted (split folders)")
        return True   # pre-extracted
    else:
        print(f"   ⚠️ Unknown structure — check contents above")
        return False

print("━" * 60)
if USE_FFPP:
    result = inspect_dataset(
        "FaceForensics++", FFPP_ROOT,
        raw_markers=["original_sequences", "manipulated_sequences"]
    )
    if result is not None:
        FFPP_IS_PRE_EXTRACTED = result
    else:
        USE_FFPP = False
else:
    print("⏭️ FaceForensics++ — skipped (USE_FFPP = False)")

print("\n" + "━" * 60)
if USE_CELEBDF:
    result = inspect_dataset(
        "Celeb-DF v2", CELEBDF_ROOT,
        raw_markers=["Celeb-real", "Celeb-synthesis"]
    )
    if result is not None:
        CELEBDF_IS_PRE_EXTRACTED = result
    else:
        USE_CELEBDF = False
else:
    print("⏭️ Celeb-DF v2 — skipped (USE_CELEBDF = False)")

print("\n" + "━" * 60)
if not USE_FFPP and not USE_CELEBDF:
    print("\n❌ No datasets found! Available in /kaggle/input/:")
    if os.path.isdir("/kaggle/input"):
        for d in sorted(os.listdir("/kaggle/input")):
            print(f"   📁 /kaggle/input/{d}")
    print("\n→ Update the paths above to match.")
else:
    print(f"\n📋 Pipeline will use: ", end="")
    parts = []
    if USE_FFPP: parts.append("FF++")
    if USE_CELEBDF: parts.append("Celeb-DF v2")
    print(" + ".join(parts))
````

## Step 3a — Extract Frames from FaceForensics++

Extracts 3 frames per video at timestamps **0.5s, 1.0s, 1.5s**.

Uses the official FF++ train/val/test split files if available.

> Auto-skipped if FF++ is disabled or pre-extracted.

````python
FRAMES_DIR = os.path.join(PROJECT_ROOT, "datasets", "frames")
os.makedirs(FRAMES_DIR, exist_ok=True)

if USE_FFPP and not FFPP_IS_PRE_EXTRACTED:
    from datasets.frame_extractor import build_split_structure
    
    print("🎬 Extracting frames from FF++ videos...")
    print(f"   Source: {FFPP_ROOT}")
    print(f"   Output: {FRAMES_DIR}")
    print(f"   Compression: c23")
    print()
    
    build_split_structure(
        ffpp_root=FFPP_ROOT,
        output_root=FRAMES_DIR,
        timestamps=[0.5, 1.0, 1.5],
        compression="c23",
        num_workers=2,
        metadata_csv=FFPP_METADATA_CSV,
    )
    
    print("\n📊 FF++ extraction results:")
    for split in ["train", "val", "test"]:
        for label in ["real", "fake"]:
            d = os.path.join(FRAMES_DIR, split, label)
            count = len(os.listdir(d)) if os.path.isdir(d) else 0
            print(f"   {split}/{label}: {count:,} frames")
    
    print("\n✅ FF++ frame extraction complete!")
elif USE_FFPP and FFPP_IS_PRE_EXTRACTED:
    print("🔗 Linking pre-extracted FF++ into project structure...")
    for split in ["train", "val", "test"]:
        for label in ["real", "fake"]:
            src = os.path.join(FFPP_ROOT, split, label)
            dst = os.path.join(FRAMES_DIR, split, label)
            if os.path.exists(src) and not os.path.exists(dst):
                os.makedirs(os.path.dirname(dst), exist_ok=True)
                os.symlink(src, dst)
                count = len(os.listdir(src))
                print(f"   ✅ {split}/{label}: {count:,} images → linked")
    print("✅ FF++ linked!")
else:
    print("⏭️ FF++ skipped")
````

## Step 3b — Extract Frames from Celeb-DF v2

Extracts 3 frames per video at timestamps **0.5s, 1.0s, 1.5s**.

Celeb-DF real = `Celeb-real/` + `YouTube-real/`, fake = `Celeb-synthesis/`.

Frames are merged into the **same** `datasets/frames/` directory alongside FF++ frames.

> Auto-skipped if Celeb-DF is disabled or pre-extracted.

````python
if USE_CELEBDF and not CELEBDF_IS_PRE_EXTRACTED:
    from datasets.frame_extractor import extract_celebdf_frames
    
    # Celeb-DF doesn't have official train/val/test splits like FF++,
    # so we extract to a staging area, then split 70/15/15
    CELEBDF_STAGING = os.path.join(PROJECT_ROOT, "datasets", "celebdf_staging")
    
    print("🎬 Extracting frames from Celeb-DF v2 videos...")
    print(f"   Source: {CELEBDF_ROOT}")
    print(f"   Staging: {CELEBDF_STAGING}")
    print()
    
    extract_celebdf_frames(
        celebdf_root=CELEBDF_ROOT,
        output_root=CELEBDF_STAGING,
        timestamps=[0.5, 1.0, 1.5],
        num_workers=2,
        metadata_csv=FFPP_METADATA_CSV,
    )
    
    # Count extracted frames
    for label in ["real", "fake"]:
        d = os.path.join(CELEBDF_STAGING, label)
        count = len(os.listdir(d)) if os.path.isdir(d) else 0
        print(f"   {label}: {count:,} frames")
    
    # Split into train/val/test (70/15/15) and merge into main frames dir
    import glob
    import random
    random.seed(42)
    
    print("\n📂 Splitting Celeb-DF into train/val/test (70/15/15)...")
    for label in ["real", "fake"]:
        src_dir = os.path.join(CELEBDF_STAGING, label)
        if not os.path.isdir(src_dir):
            continue
        
        files = sorted(glob.glob(os.path.join(src_dir, "*.*")))
        random.shuffle(files)
        
        n = len(files)
        n_train = int(0.70 * n)
        n_val   = int(0.15 * n)
        
        splits = {
            "train": files[:n_train],
            "val":   files[n_train:n_train + n_val],
            "test":  files[n_train + n_val:],
        }
        
        for split_name, split_files in splits.items():
            dst_dir = os.path.join(FRAMES_DIR, split_name, label)
            os.makedirs(dst_dir, exist_ok=True)
            for f in split_files:
                # Prefix with 'celebdf_' to avoid name clashes with FF++ frames
                dst = os.path.join(dst_dir, "celebdf_" + os.path.basename(f))
                if not os.path.exists(dst):
                    os.rename(f, dst)
            print(f"   {split_name}/{label}: {len(split_files):,} → moved")
    
    # Cleanup staging
    import shutil
    if os.path.isdir(CELEBDF_STAGING):
        shutil.rmtree(CELEBDF_STAGING)
    
    print("\n✅ Celeb-DF v2 extraction & split complete!")

elif USE_CELEBDF and CELEBDF_IS_PRE_EXTRACTED:
    print("🔗 Linking pre-extracted Celeb-DF v2...")
    for split in ["train", "val", "test"]:
        for label in ["real", "fake"]:
            src = os.path.join(CELEBDF_ROOT, split, label)
            dst = os.path.join(FRAMES_DIR, split, label)
            if os.path.exists(src):
                # Can't symlink if FF++ already created the dir, so copy instead
                os.makedirs(dst, exist_ok=True)
                import glob
                for f in glob.glob(os.path.join(src, "*.*")):
                    dst_file = os.path.join(dst, "celebdf_" + os.path.basename(f))
                    if not os.path.exists(dst_file):
                        os.symlink(f, dst_file)
                count = len(glob.glob(os.path.join(src, "*.*")))
                print(f"   ✅ {split}/{label}: {count:,} images → linked")
    print("✅ Celeb-DF v2 linked!")
else:
    print("⏭️ Celeb-DF v2 skipped")
````

````python
# Combined dataset summary
print("📊 Combined dataset summary:")
print("━" * 40)
total = 0
for split in ["train", "val", "test"]:
    for label in ["real", "fake"]:
        d = os.path.join(FRAMES_DIR, split, label)
        count = len(os.listdir(d)) if os.path.isdir(d) else 0
        total += count
        print(f"   {split:5s}/{label:4s}: {count:>8,} images")
print("━" * 40)
print(f"   Total:        {total:>8,} images")
````

## Step 4 — Face Detection & Cropping (MTCNN)

Detects faces in each frame using MTCNN and crops to **224×224**.

> Auto-skipped if both datasets are pre-extracted.

````python
need_cropping = (USE_FFPP and not FFPP_IS_PRE_EXTRACTED) or (USE_CELEBDF and not CELEBDF_IS_PRE_EXTRACTED)

if need_cropping:
    from datasets.face_cropper import process_dataset_splits
    
    print("👤 Detecting & cropping faces with MTCNN...")
    print(f"   Input:  {FRAMES_DIR}")
    print(f"   Output: {FRAMES_DIR} (in-place)")
    print(f"   Size:   224×224")
    print()
    
    device = "cuda" if torch.cuda.is_available() else "cpu"
    
    process_dataset_splits(
        frames_root=FRAMES_DIR,
        output_root=FRAMES_DIR,
        face_size=224,
        margin=20,
        device=device,
    )
    
    print("\n📊 Final dataset after face cropping:")
    for split in ["train", "val", "test"]:
        for label in ["real", "fake"]:
            d = os.path.join(FRAMES_DIR, split, label)
            count = len(os.listdir(d)) if os.path.isdir(d) else 0
            print(f"   {split}/{label}: {count:,} face images")
    
    print("\n✅ Face preprocessing complete!")
else:
    print("⏭️ Face cropping skipped — datasets are pre-extracted")
````

## Step 5 — Train the Multi-Branch Detector

Architecture: **EfficientNet-B4 (spatial)** + **FFT-CNN (frequency)** + **SRM-CNN (noise)**

Training features:
- Mixed precision (AMP) for faster training
- Gradient accumulation for effective batch size of 32
- Cosine annealing LR schedule
- Automatic checkpoint saving every epoch
- Resume support if session times out

````python
from utils.config import Config

cfg = Config()

# ╔══════════════════════════════════════════════════════════════════╗
# ║                    TRAINING CONFIGURATION                       ║
# ╚══════════════════════════════════════════════════════════════════╝

# Paths — override for Kaggle
cfg.checkpoints_dir = "/kaggle/working/checkpoints"
cfg.outputs_dir     = "/kaggle/working/outputs"

# CSV overrides
cfg.train_csv = "/kaggle/input/my-dataset/train.csv" # Replace with actual csv path
cfg.val_csv = "/kaggle/input/my-dataset/val.csv"     # Replace with actual csv path
os.makedirs(cfg.checkpoints_dir, exist_ok=True)
os.makedirs(cfg.outputs_dir, exist_ok=True)

# Hyperparameters — tuned for Kaggle P100 (16 GB VRAM)
cfg.batch_size                  = 16    # Safe for P100; try 24 if VRAM allows
cfg.num_epochs                  = 25
cfg.learning_rate               = 1e-4
cfg.weight_decay                = 1e-2
cfg.num_workers                 = 2     # Kaggle has 4 CPUs total
cfg.use_amp                     = True  # Always use mixed precision
cfg.gradient_accumulation_steps = 2     # Effective batch = 16 × 2 = 32

# Seed for reproducibility
cfg.set_seed()

# Print full config
print(cfg)
````

## 💡 Troubleshooting

### Session timed out mid-training?
Change `RESUME_PATH` in Step 5 to:
```python
RESUME_PATH = "/kaggle/working/checkpoints/last_checkpoint.pt"
```
Then re-run from Step 5 onward. All progress is preserved.

### Out of GPU memory?
Reduce `cfg.batch_size` to `8` and increase `cfg.gradient_accumulation_steps` to `4`.
Effective batch size stays at 32.

### No FF++ split files?
The extractor will fall back to a flat structure. You'll need to manually split the data
into `train/real`, `train/fake`, `val/real`, `val/fake`, `test/real`, `test/fake`.
